This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  3


Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to  0.2 and random_state to 2. The training data and test data should be assigned to the following labels.


<code>X_train, X_test, Y_train, Y_test</code>

## Authors


```python
from sklearn.metrics import (
    classification_report, roc_auc_score, accuracy_score
)

def get_metrics(model, X_test, Y_test, name="Model"):
    y_pred = model.predict(X_test)
    try:
        y_score = model.predict_proba(X_test)[:,1]
    except AttributeError:
        y_score = y_pred  # fallback for models without predict_proba

    acc = accuracy_score(Y_test, y_pred)
    report = classification_report(Y_test, y_pred, output_dict=True)
    precision = report["1"]["precision"]
    recall = report["1"]["recall"]
    f1 = report["1"]["f1-score"]
    auc = roc_auc_score(Y_test, y_score)
    return {
        "Model": name,
        "Accuracy": round(acc, 2),
        "F1 Score": round(f1, 2),
        "ROC AUC": round(auc, 2),
        "Precision": round(precision, 2),
        "Recall": round(recall, 2)
    }

# Collect results for each model
results = []
results.append(get_metrics(logreg_cv, X_test, Y_test, "Logistic Regression"))
results.append(get_metrics(svm_cv, X_test, Y_test, "SVM"))
results.append(get_metrics(tree_cv, X_test, Y_test, "Decision Tree"))
results.append(get_metrics(knn_cv, X_test, Y_test, "kNN"))

# Make DataFrame and add ranking
df_results = pd.DataFrame(results)
df_results["Rank"] = df_results["Accuracy"].rank(ascending=False, method="min").astype(int)
df_results = df_results.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)

print(df_results.to_markdown(index=False))
```
